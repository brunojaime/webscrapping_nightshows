```python
def reemplazar_caracteres(text_in,caracteres):
     for caracter in caracteres:
          for entrada in caracter['in']:
               if entrada in text_in:
                    text_in = text_in.replace(entrada,caracter['out'])
     print(text_in)
     return text_in

def validar_url(url_general, url_especifica):
     while (url_especifica!=""):
          url_completa = f"{url_general}{url_especifica}"
          if (requests.get(url_completa).status_code==200):
               return url_completa
          else:
               palabras = url_especifica.split("-")[:-1]
               url_especifica = '-'.join(palabras)
               print(url_especifica)
     return 0
```

```python
import unicodedata

def obtener_url_final(url_entrada):
    
    url_entrada = url_entrada.strip().lower()
    caracteres_invalidos= [":",".",","]
    url_sin_acentos = ''.join((c for c in unicodedata.normalize('NFD', url_entrada)
                             if unicodedata.category(c) != 'Mn'))

    caracteres_especiales =[

        {"in":[" "],"out": "-"}
        ]
    
    url_final= reemplazar_caracteres(url_sin_acentos,caracteres_especiales)
    url_final =validar_url(url, url_final)   
    return url_final
```

```python
# Asumiendo que `html` es tu documento HTML ya cargado como string
import re
from datetime import datetime

soup = bs(html.text, 'html.parser')
def obtener_fecha(fecha_str):
    fecha_upper = fecha_str.upper()
    dia = obtener_dia(fecha_upper)
    mes = obtener_mes(fecha_upper)
    anio = datetime.now().year
    return f"{dia}/{mes}/{anio}"



def obtener_dia(fecha_str):
    numeros = re.findall(r'\d+',fecha_str)
    if numeros:
        return numeros[0]
    else:
        return None
def obtener_mes(fecha_str):
    meses = {
        'ENERO': 1, 'FEBRERO': 2, 'MARZO': 3, 'ABRIL': 4, 'MAYO': 5, 'JUNIO': 6,
        'JULIO': 7, 'AGOSTO': 8, 'SEPTIEMBRE': 9, 'OCTUBRE': 10, 'NOVIEMBRE': 11, 'DICIEMBRE': 12
    }
    
    for palabra in fecha_str.split():
        if palabra in meses:
            return  meses[palabra]
    
    return None  
# Encuentra todos los divs que contienen la clase "espectaculo"
espectaculos = soup.find_all('div', class_='espectaculo')
espectaculos_list = [
    {
        "categoria":espectaculo.find('div',class_= 'categoria').text.strip(),
        'nombre' :espectaculo.find('div',class_= 'nombre').text.strip().upper(),
        'fecha': obtener_fecha(espectaculo.find('p').text.strip()),
        "hora": espectaculo.find_all('p')[1].text.strip(),
        "url": obtener_url_final(espectaculo.find('div',class_= 'nombre').text)         
    }
    for espectaculo in espectaculos
]

espectaculos_list


```

```python
def imprimir_precio(soup):
    try:
        show = soup.find_all('li',class_='dato_entradas')

        for sho in show:
                este_show = sho.find_all('p')
                return(este_show)
    except:
         return "No se encontr√≥"
```

```python
espec = espectaculos_list
for espectaculo in espec: 
    html = requests.get(espectaculo['url'])
    soup = bs(html.content,"html.parser")
    print(imprimir_precio(soup))
    
```
